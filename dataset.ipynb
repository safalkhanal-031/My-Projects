```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# Get webpage
url = "https://evostore.com.np/apple/Mac/Buy_MacBook-Air_in_Nepal"
webpage = requests.get(url).text
```

```python
# Parse HTML
soup = BeautifulSoup(webpage, "lxml")
```

```python
# Lists to store data
macbooks = []
volume = []
price = []
```

```python
# Find all laptop blocks
laptops = soup.find_all("div", class_="text")
```

```python

for laptop in laptops:
    macbooks.append(laptop.find("div", class_="name").text.strip())
    volume.append(laptop.find("div", class_="volume").text.strip())
    
    rawprice = laptop.find("div", class_="price").text.strip()
    price.append(rawprice.split(" NPR")[0])  # take only first price
```

```python
# Check lengths
print(len(macbooks), len(volume), len(price))
```

```python
# Create DataFrame
df = pd.DataFrame({
    "Name": macbooks,
    "Conf": volume,
    "Price": price
})
```
