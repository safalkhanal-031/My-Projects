```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
for i in range(1, 257):
    url = f"https://www.nepalhomes.com/search?find_property_category=5d662c7b8f12c7035cd39315&page={i}&sort=1"
    html = requests.get(url)
    with open(f"htmls/page{i}.html", "w", encoding="utf-8") as f:
        f.write(html.text)
        print(f"Downloaded page {i} successfully")
```

```python
f.close()
```

```python
PRICE = []
LOCATION = []
ROAD_ACCESS = []
FACING_DIRECTION = []
LAND_AREA = []

for i in range(1, 257):
    with open(f"htmls/page{i}.html", "r") as f:
        soup = BeautifulSoup(f.read(), "html.parser")

        data = soup.find_all("div", class_="property__card-text")

        for d in data:
            PRICE.append(d.find("span", class_="price-tag tag tag-lg tag-gray-primary").text.strip())
            LOCATION.append(d.find("p", class_="property__card-location").text.strip())

            found_area = False
            found_road = False
            found_direction = False

            specs = d.find_all("div", class_="spec-item")
            for s in specs:
                p = s.find("p")
                if p:
                    text = p.text.strip()

                    if "aana" in text or "kattha" in text:
                        LAND_AREA.append(text)
                        found_area = True
                    elif "Feet" in text or "Meter" in text:
                        ROAD_ACCESS.append(text)
                        found_road = True
                    else:
                        FACING_DIRECTION.append(text)
                        found_direction = True

            # append None if not found
            if not found_area:
                LAND_AREA.append(None)
            if not found_road:
                ROAD_ACCESS.append(None)
            if not found_direction:
                FACING_DIRECTION.append(None)
```

```python
print(len(PRICE), len(LOCATION), len(ROAD_ACCESS), len(FACING_DIRECTION), len(LAND_AREA))
```

```python
df = pd.DataFrame({
    "Price": PRICE,
    "Location": LOCATION,
    "Road_Access": ROAD_ACCESS,
    "Facing_Direction": FACING_DIRECTION,
    "Land_AREA": LAND_AREA
})
```

```python
df.to_csv("nepali_land_data.csv", index=False)
```
